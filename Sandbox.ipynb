### Transformar pdf de matrículas deferidas em DF

```python
import pdfplumber
import pandas as pd
```

```python
len(pdf.pages)
```

```python
pdf = pdfplumber.open('matriculas_deferidas_pos_ajuste2022_2.pdf')

it=0

for i in range(len(pdf.pages)):
    
    text = pdf.pages[i].extract_text()
    
    text_trat = text.split('\n')
    
    text_trat.pop(0)
    text_trat.pop(0)
    text_trat.pop(0)
    text_trat.pop(len(text_trat)-1)
    
    list_text_trat = []
    
    for a in text_trat:
        list_text_trat.append(a.split(' ',2))
        
    pd.DataFrame(list_text_trat,columns=['RA','TURMA','TURMA_NOME']).to_csv('matriculas_deferidas_pos_ajuste2022_2.csv',header=(it==0),index=False,mode='a',sep=';',encoding='utf-8')
    
    print('página {} de {} processada'.format(i,len(pdf.pages)))
    
    it = it+1
    
pdf.close()
```
